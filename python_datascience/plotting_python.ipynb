#### Matplotlib

- Basic plotting package 
- Very customisable - every single line/pixel 
- Need alot of code to write a simple plot

```python
# Make x & y variables using numpy 
x = np.arange(0, 5, 0.1)
y = np.sin(x)

print('This is x:')
print(x)

print()

print('This is y:')
print(y)
```

### Matplot lib object orientated interface

```python

fig = plt.figure()   # create figure object - contain axes, graphics, text
ax = plt.axes()      # create axes - a single pane

ax.plot(x, y) # create continuous line plot

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)
ax.set_title('title of plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

### Matplot lib Pyplot interface

```python
plt.plot(x, y)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.title('title of plt')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### Scatterplot

```python
tips = sns.load_dataset("tips")
display(tips)

# set color to Yes/No depending on value in 'smoker column' 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            data=tips)

# as above plus shape of points for time column
sns.relplot(x="total_bill",
            y="tip",
            hue="smoker",
            style="time",
            data=tips);
```

```python

# col = Time -> create 2 plots (facets) split by time column 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            col="time", 
            data=tips)
```

#### Pairs plot

```python
iris = sns.load_dataset("iris")
display(iris.head())

# plot pairs plot
# assign plot to object 'g' - sets up grid
g = sns.PairGrid(iris, hue="species")

# plot histogram on diagonal
g.map_diag(plt.hist) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend();
```

### Lineplot example

```python
fmri = sns.load_dataset("fmri")
display(fmri.head())
sns.relplot(x="timepoint",
            y="signal", 
            hue="event", 
            kind="line", 
            data=fmri)

sns.relplot(x="timepoint", 
            y="signal", 
            hue="region", 
            style="event",
            dashes=False, 
            markers=True, 
            kind="line", 
            data=fmri);
```

### Facets

```python
sns.relplot(x="timepoint",
            y="signal",
            hue="subject",
            col="region",
            row="event",
            height=3,
            kind="line",
            estimator=None,
            data=fmri)
```

### Plot Categorical Data

```python
# create violin plots
g = sns.catplot(x="day", 
                y="total_bill", 
                kind="violin",
                inner=None,
                data=tips)

# add individual datapoints -> provide ax argument
sns.swarmplot(x="day",
              y="total_bill",
              color="k",
              size=3,
              data=tips,
              ax=g.ax) 
```

Rename columns and remove unwanted columns (Chr, Start, End, Strand, Length) 


```python
counts = df.drop(columns = ['Chr', 'Start', 'End', 'Strand', 'Length'], axis=1)
counts.columns = counts.columns.str.replace('sam/', '').str.replace('.bam','')

#we can then use the df.rename way, here the new names are explicitly mentioned 
counts.rename(columns={'ERR1755082': 'cd4_r1', 'ERR1755083': 'cd4_r2', 'ERR1755084': 'cd4_r3', 'ERR1755085': 'cd8_r1', 'ERR1755086': 'cd8_r2', 'ERR1755087': 'cd8_r3'}, inplace = True)
counts
```

Filter out genes with no counts


```python
counts['gene_count'] = counts.sum(axis=1) #to be able to filter it, we first calculate the sum across all samples for each gene, so if this = 0, we can then filter it out
counts
```

```python
counts = counts.set_index(['Geneid'])
counts
```

```python
filtered_counts = counts[counts['gene_count'] > 0]
filtered_counts #if you only state the name, this is the same as display
```

Plot scatter plot of correlation of replicates (pairs plot) (you will need to google how to show plots in jupyter notebook)

```python
filtered_counts = counts[counts.gene_count>0].drop(columns='gene_count') #Jian's elegant version to filter and then delete the column used for filtering at the same time
filtered_counts.head(10)
```

```python
#plot pairs plot
#
g = sns.PairGrid(np.log2(filtered_counts + 1)) #problem is that this log transform doesn't work with 0 values, so we need to add + 1 to make it work; this puts all data on a scale so it can be compared

# plot histogram on diagonal
g.map_diag(plt.hist, bins = 100) #this defines the number of bins
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend()
```

Identify gene with highest expression in each sample


```python
filtered_counts_gene_index = filtered_counts.set_index('Geneid')
maxValueIndexObj = filtered_counts_gene_index.idxmax()
maxValueIndexObj 
```

```python
#to check that those genes are the ones with the highest value
filtered_counts[filtered_counts['Geneid'].isin(['l7Rn6', 'Eef1a1', 'Actb'])]
```

Normalise read counts to total reads per sample & convert to 'counts per million (CPM)'


```python
sum_counts = filtered_counts_gene_index.sum() #to sum up the reads per sample, best to used the indexed dataframe here #to divide the dataframe by the sum of the columns
display(filtered_counts_gene_index)
display(sum_counts)
filtered_counts_CPM = filtered_counts_gene_index.div(sum_counts, axis = 1) * 1e6 #to normalise for counts per million
filtered_counts_CPM
```

Log transform normalised read counts


```python
log2_filtered_counts_CPM = np.log2(filtered_counts_CPM + 1) #+1 is necessary again, so it is happy with 0 values
log2_filtered_counts_CPM
```

Create new dataframe of 100 genes with highest average variance in log2 cpm across conditions

```python
log2_filtered_counts_CPM['variance'] = log2_filtered_counts_CPM.var(axis = 1)
log2_filtered_counts_CPM
top_100_variance = log2_filtered_counts_CPM.sort_values(by = ['variance'], ascending = False).head(100)
top_100_variance
```

Plot the of expression of the top 5 most variable genes (log cpm) across all samples (line plot)


```python
#space and backslash allow to move the commands to the next line, so it's more readable 
top_5_variance_melt = top_100_variance.head(5) \
                                      .drop('variance', axis = 1) \
                                      .reset_index() \
                                      .melt(id_vars = ['Geneid'], var_name = 'sample', value_name = 'counts') 
top_5_variance_melt

sns.lineplot(data=top_5_variance_melt, x="Geneid", y="counts", hue="sample")
plt.show() #this is required if you don't want both graphs to be displayed in the same figure
sns.lineplot(data=top_5_variance_melt, x="sample", y="counts", hue="Geneid")
```

Plot a heatmap (here: cluster map) for top 100 most variable genes (log cpm) across all samples


```python
g = sns.clustermap(top_100_variance.drop('variance', axis = 1), figsize = (10,30), cmap="mako")
```
